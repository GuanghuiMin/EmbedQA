```python
import torch.nn as nn


class WordEmbedding(nn.Module):
    '''
    In : (N, sentence_len)
    Out: (N, sentence_len, embd_size)
    '''
    def __init__(self, vocab_size_w, w_embd_size, pre_embd_w, is_train_embd=False):
        super(WordEmbedding, self).__init__()
        self.embedding = nn.Embedding(vocab_size_w, w_embd_size)
        if pre_embd_w is not None:
            self.embedding.weight = nn.Parameter(pre_embd_w, requires_grad=is_train_embd)

    def forward(self, x):
        return self.embedding(x)
```

```python
import torch
import numpy as np
vocab_size_w = 100
w_embd_size = 20
pre_embd_w = torch.randn(100, 20)
```
